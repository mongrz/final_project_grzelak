```python
def replace_from_dict(string, dictionary):
  for k in dictionary.keys():
    string = string.replace(k, dictionary[k])
  return string
```

```python
data['created_at'] = data.apply(lambda row: replace_from_dict(row['created_at'], weekday_name_dict), axis = 1)
```

```python
import spacy
```

```python
nlp = spacy.load("pl_core_news_sm")
```

```python
def remove_stops(text):
  doc = nlp(text)
  filtered_words = [token.text for token in doc if not token.is_stop]
  clean_text = ' '.join(filtered_words)
  return clean_text
```

```python
data['text_witout_stopwords'] = data.apply(lambda row: remove_stops(row['text']), axis = 1)
```

```python
#Extract people (persName) and add them to a column called persons
def extract_entities(text, ent_type):
  doc = nlp(text)
  entities = []
  for ent in doc.ents:
    if ent.label_ == ent_type:
      entities.append(ent)
  return entities
```

```python
data['persons'] = data.apply(lambda row: extract_entities(row['text'], 'persName'), axis = 1)
```

```python
data.sample(25)
```

```python
#Extract places (placeName) and add them to a column called places
data['places'] = data.apply(lambda row: extract_entities(row['text'], 'placeName'), axis = 1)
```

```python
#Extract organisations (orgName) and place them in the column called organisations
data['organisations'] = data.apply(lambda row: extract_entities(row['text'], 'orgName'), axis = 1)
```

```python
#Using the matplotlib package, create a graph showing the number of tweets per day of the week.
weekday_sorting_dict = {"Monday": 0, "Tuesday":1, "Wednesday":2, "Thursday": 3, "Friday":4, "Saturday": 5, "Sunday": 6}

def particular_sort(input):
  return input.map(lambda x : weekday_sorting_dict.get(x))

```

```python
import matplotlib.pyplot as plt

ax = data['weekday'].value_counts().sort_index(key = lambda x: particular_sort(x)).plot(kind='bar',
                                    figsize=(14,8),
                                    title="Number of tweets per day of the week")
ax.set_xlabel("Days of the week")
ax.set_ylabel("Count")

plt.show()
```
